```python
import pandas as pd
import json

def parse_jsonl_file(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            data.append(json.loads(line))
    return data

def jsonl_to_dataframe(file_path):
    json_data = parse_jsonl_file(file_path)
    df = pd.DataFrame(json_data)
    return df

file_path = './crag_task_3_v2/crag_task_3_v2_1.jsonl'
#data = parse_jsonl_file(file_path)
```

```python
from bs4 import BeautifulSoup

def html_to_text(html_string):
    soup = BeautifulSoup(html_string, 'html.parser')
    text = soup.get_text()
    return text
```

```python
df = jsonl_to_dataframe(file_path)
#df.head(10)
```

```python
#pages = df.iloc[4].search_results[:5] 
```

```python
def clean_HTML(pages):
    search_res_refined = []
    for page_json in pages[:5]:
        plain_text = html_to_text(page_json['page_result'])
        del page_json['page_result']
        page_json['plain_text'] = plain_text
        search_res_refined.append(page_json)
    return search_res_refined
```

```python
# clean_HTML(df.iloc[4].search_results[:5] )
```

```python
df['search_results'] = df['search_results'].apply(lambda x: clean_HTML(x))
```
